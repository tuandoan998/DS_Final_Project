```python
from requests_html import HTMLSession
import pandas as pd
```

```python
session = HTMLSession()
```

```python
def get_content(url):
    r = session.get(url)
    try:
        body = r.html.find('.story-body', first=True)
        title = body.find('h1', first=True).text
    except:
        return
    
    h2 = ''
    content = ''
    inner = body.find('.story-body__inner', first=True)
    h2_html = inner.find('h2')
    for i in h2_html:
        h2 += i.text
    texts = inner.find('p')
    for i in texts:
        content+=i.text
    
    return title, h2, content
```

```python
business_urls, entertainment_urls, science_urls, sport_urls, technology_urls = [], [], [], [], []
with open("urls/business.txt", "r") as text_file:
    business_urls = [line.rstrip('\n') for line in text_file]
    text_file.close()
with open("urls/entertainment.txt", "r") as text_file:
    entertainment_urls = [line.rstrip('\n') for line in text_file]
    text_file.close()
with open("urls/science.txt", "r") as text_file:
    science_urls = [line.rstrip('\n') for line in text_file]
    text_file.close()
with open("urls/sport.txt", "r") as text_file:
    sport_urls = [line.rstrip('\n') for line in text_file]
    text_file.close()
with open("urls/technology.txt", "r") as text_file:
    technology_urls = [line.rstrip('\n') for line in text_file]
    text_file.close()
```

```python
def parse_bbcnews(urls, category):
    df = pd.DataFrame(columns=["label", "title", "h2", "content"])
    for url in urls:
        ret_value = get_content(url)
        if ret_value is not None:
            title, h2, content = ret_value
            df = df.append({
                 "label": category,
                 "title": title,
                 "h2": h2,
                 "content": content}, ignore_index=True)
            print(title)
    return df
```

```python
df = parse_bbcnews(business_urls, "business")
df.to_csv('business.csv', sep='\t', index=False)
```

```python
df = parse_bbcnews(entertainment_urls, "entertainment")
df.to_csv('entertainment.csv', sep='\t', index=False)
```

```python
df = parse_bbcnews(science_urls, "science")
df.to_csv('science.csv', sep='\t', index=False)
```

```python
df = parse_bbcnews(technology_urls, "technology")
df.to_csv('technology.csv', sep='\t', index=False)
```

```python
def get_content_sport(url):
    r = session.get(url)
    try:
        article = r.html.find('article', first=True)
        title = article.find('h1', first=True).text
    except:
        return
    
    h2 = ''
    content = ''
    inner = article.find('#story-body', first=True)
    h2_html = inner.find('h3')
    for i in h2_html:
        h2 += i.text
    texts = inner.find('p')
    for i in texts:
        content+=i.text
    
    return title, h2, content
```

```python
df = pd.DataFrame(columns=["label", "title", "h2", "content"])
for url in sport_urls:
    ret_value = get_content_sport(url)
    if ret_value is not None:
        title, h2, content = ret_value
        df = df.append({
                 "label": 'sport',
                 "title": title,
                 "h2": h2,
                 "content": content}, ignore_index=True)
        print(title)
df.to_csv('sport.csv', sep='\t', index=False)
```
